```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings(action='ignore')

import requests 
import json 
import time 
from datetime import datetime, timedelta
from bs4 import BeautifulSoup
from scipy.stats import trim_mean

pd.options.display.max_rows=100
pd.options.display.max_columns=100
```

```python
url = 'https://api.bunjang.co.kr/api/1/find_v2.json?q={}&order=score&page={}&request_id=20211110215050&stat_device=w&n=100&stat_category_required=1&req_ref=search&version=4'
keywords = ['나이키 덩크로우']
```

```python
def dunk_crawler(keyword, n=85):
    df = pd.DataFrame()
    
    for i in range(0,n):

        seed = np.random.randint(100)
        np.random.seed(seed)
        a = np.random.randint(3)
        time.sleep(a)
        
        url_formating = url.format(keyword, i) # url에 포맷팅을 적용한다.
        info = {'referer' : 'https://m.bunjang.co.kr/',
           'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36'}
         
       
        print(f'{i}th, {(i+1)*100}개째 상품을 크롤링 중')
       
        resp = requests.get(url_formating, headers = info)

        
        if resp.status_code == requests.codes.ok:
            data = json.loads(resp.text)
            next_df = pd.DataFrame(data['list'])
            df = pd.concat([df, next_df])

        
        else:
            print(f'{i}번째 요청이 잘못되었습니다.')
            pass 

    
    print('크롤링 완료!!')
    result = df.reset_index(drop=True)
    return result
```

```python
dunk_df = dunk_crawler(keywords[0] , 85) 
```

```python
dunk_df.head(3)
```

```python
columns = ["location", "name","num_faved", "pid", "price","status","uid","update_time","used"]
dunk_df= dunk_df.loc[: ,columns]

dunk_df.head(3)
```

```python
def get_date(x):
    return datetime(1970,1,1,9) + timedelta(seconds=int(x))
```

```python
def datetime_maker(df):
    df['datetime'] = df['update_time'].apply(get_date)
    return df
```

```python
dunk_df = datetime_maker(dunk_df)
dunk_df.head()
```

```python
### 시간,분,초 삭제하고 새로운 date 컬럼 생성 

def dt_index(x):
    return str(x)[:10]
```

```python
dunk_df["date"] = dunk_df["datetime"].apply(dt_index)
dunk_df.head(3)
```

```python
dunk_df["date"] = pd.to_datetime(dunk_df["date"]) # date 컬럼 데이터타입을 datetime으로 변경 
```
